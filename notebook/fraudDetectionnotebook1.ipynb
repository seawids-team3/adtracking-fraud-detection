```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline
```

```python
data = 'train.csv'

trainData = pd.read_csv(data)
```

```python
trainData
```

```python
plt.figure(figsize=(12,10))
cor = trainData.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
plt.figure(figsize=(12,10))
cor = trainData.query('is_attributed ==1').corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
groupByIP = trainData.groupby('app')['ip'].count().reset_index()
```

```python
plt.figure(figsize=(8,7))
plt.plot(groupByIP['app'], groupByIP['ip'])

```

```python
groupbyChannelId = trainData.groupby('channel')['ip'].count().plot()
plt.ylabel('no of ips')
plt.title('number of networks based on channel')
```
